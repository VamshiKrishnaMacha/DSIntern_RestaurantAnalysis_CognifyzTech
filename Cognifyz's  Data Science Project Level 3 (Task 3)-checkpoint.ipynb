## Task: Data Visualization

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('Dataset.csv')


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot a histogram to visualize the distribution of ratings
plt.figure(figsize=(10, 6))
sns.histplot(data['Aggregate rating'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Group by cuisine and calculate the average rating for each cuisine
average_ratings_cuisine = data.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

# Plot a bar chart to compare the average ratings of different cuisines
plt.figure(figsize=(12, 6))
average_ratings_cuisine.plot(kind='bar', color='salmon')
plt.title('Top 10 Cuisines with Highest Average Ratings')
plt.xlabel('Cuisine')
plt.ylabel('Average Rating')
plt.show()
```

```python
# Group by city and calculate the average rating for each city
average_ratings_city = data.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

# Plot a bar chart to compare the average ratings of different cities
plt.figure(figsize=(12, 6))
average_ratings_city.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Cities with Highest Average Ratings')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.show()
```
